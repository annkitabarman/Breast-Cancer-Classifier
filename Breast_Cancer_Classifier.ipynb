Importing the dependencies

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
data = pd.read_csv('data.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.isnull().any()
```

```python
data.info()
```

```python
data = data.drop('Unnamed: 32', axis=1)
```

```python
data.describe()
```

```python
data.diagnosis.replace(['M', 'B'], [1,0], inplace=True)
```

```python
data.head()
```

```python
data['diagnosis'].value_counts()
```

```python
data.groupby('diagnosis').mean()
```

```python
sns.countplot('diagnosis', data=data); #1=M,0=B
plt.show()
```

Splitting the features and target

```python
X = data.drop(['id','diagnosis'],axis=1)
Y = data['diagnosis']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training and test data

```python
xTrain, xTest, yTrain, yTest = train_test_split(X,Y, train_size=0.8,stratify = Y, random_state=0)
```

```python
xTrain.shape, yTrain.shape
```

Model Training

```python
regressor = LogisticRegression(solver = 'newton-cg')
```

```python
regressor.fit(xTrain, yTrain)
```

Accuracy Score

```python
train_pred = regressor.predict(xTrain)
```

```python
train_acc = accuracy_score(yTrain, train_pred)
```

```python
print(train_acc * 100)
```

```python
test_pred = regressor.predict(xTest)
```

```python
test_acc = accuracy_score(yTest, test_pred)
```

```python
print(test_acc * 100)
```

```python
print(confusion_matrix(yTrain, train_pred))
```

```python
print(confusion_matrix(yTest, test_pred))
```
